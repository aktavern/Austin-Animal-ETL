```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv1 = 'Resources/Austin_Animal_Center_Intakes.csv'
intake = pd.read_csv(csv1)
intake.head().sort_values('Animal ID')
```

```python
intake_df = intake[['Animal ID','Name','DateTime','Intake Type','Intake Condition','Animal Type','Age upon Intake','Breed','Color']]
intake_df.head()
```

```python
intake_df = intake_df.rename(columns ={'Animal ID': 'id',
             'Name': 'intake_name',
             'DateTime': 'intake_time',
             'Intake Type': 'intake_type',
             'Intake Condition': 'intake_condition',
             'Animal Type': 'animal_type',
             'Age upon Intake' :'intake_age',
             'Breed': 'breed',
             'Color': 'color'
             })
intake_df.head()
```

```python
outcomes = pd.read_csv('Resources/Austin_Animal_Center_Outcomes.csv')
outcomes.head().sort_values('Animal ID')
```

```python
outcomes_df= outcomes [['Animal ID','Name','DateTime','Outcome Type','Age upon Outcome']]
outcomes_df.head()
```

```python
outcomes_df = outcomes_df.rename(columns ={'Animal ID': 'id',
                                         'Name' : 'outcome_name',
                                         'DateTime': 'outcome_date',
                                         'Age upon Outcome':'outcome_age',
                                          'Outcome Type': 'outcome_type'})
outcomes_df.head()
```

```python
rds_connection_string = "postgres:421Monkey!@localhost:5432/Austin_Animal_Shelter"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
intake_df.to_sql(name = 'intakes', con = engine, if_exists='append', index =True)
```

```python
outcomes_df.to_sql(name = 'outcomes' , con = engine, if_exists ='append', index = True)
```
